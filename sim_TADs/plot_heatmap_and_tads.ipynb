```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
out_location = "yielded/"
```

```python
mtx = np.loadtxt("matrices/simHiC_countMatrix_noise4_sim1.txt")
```

```python
#basic usage
bgn = 700
end = 900
sns.heatmap(mtx[bgn:end, bgn:end])
```

```python
TADs = np.loadtxt(f"trueTADs/simHiC_TADintervals_coords_noise4_sim1.txt", ndmin=2)
res = 40000
bgn = 700
end = 900
tads_color = "yellow"
plt.figure(figsize=(24,20))
sns.heatmap(mtx[bgn:end, bgn:end], cmap='Greys_r')
plt.xticks([])
plt.yticks([])
plt.title(f"True TADs")
for i in TADs:
    tad_bgn = i[0] / res - bgn
    tad_end = i[1] / res - bgn
    plt.plot([tad_bgn, tad_end], [tad_bgn, tad_bgn], color=tads_color, linewidth=5)
    plt.plot([tad_end, tad_end], [tad_bgn, tad_end], color=tads_color, linewidth=5)
plt.savefig(f"true_tads_noise4_sim1.png")
```

```python
ideal_gamma = {"armatus": -0.3, "lava_armatus": -0.3, "lava_variance": 6, "lava_potts": 16, "lava_corner": 2, "lava_modularity":16}
ideal_params = {"ii": (7, 160000.0), "di": (13, 160000)}
method_name = {"armatus":"Armatus\nstandalone", "lava_armatus":"Armatus\nlavaburst", "lava_variance":"Variance", "lava_potts":"Potts", "lava_corner":"Corner", "lava_modularity":"Modularity", "ii":"Insulation\nscore", "di":"Directionality\nindex", "hicseg":"HiCseg N", "hicseg_P":"HiCseg P"}
```

```python
for method in ideal_gamma.keys():
    if method == "lava_variance":
        continue
    res = 40000
    TADs = np.loadtxt(f"{out_location}{method}_noise{12}_sim{1}_gamma{ideal_gamma[method]}.txt", ndmin=2)
    bgn = 700
    end = 900
    tads_color = "yellow"
    plt.figure(figsize=(24,20))
    sns.heatmap(mtx[bgn:end, bgn:end], cmap='Greys_r')
    plt.xticks([])
    plt.yticks([])
    plt.title(f"{method_name[method]} TADs")
    for i in TADs:
        tad_bgn = i[0] / res - bgn
        tad_end = i[1] / res - bgn
        plt.plot([tad_bgn, tad_end], [tad_bgn, tad_bgn], color=tads_color, linewidth=5)
        plt.plot([tad_end, tad_end], [tad_bgn, tad_end], color=tads_color, linewidth=5)
    plt.savefig(f"{method}_tads_noise12_sim1.png")
```

```python
for method in ideal_params.keys():
    res = 40000
    TADs = np.loadtxt(f"{out_location}{method}_noise{12}_sim{1}_window{ideal_params[method][1]}_cutoff{ideal_params[method][0]}.txt", ndmin=2)
    bgn = 700
    end = 900
    tads_color = "yellow"
    plt.figure(figsize=(24,20))
    sns.heatmap(mtx[bgn:end, bgn:end], cmap='Greys_r')
    plt.xticks([])
    plt.yticks([])
    plt.title(f"{method_name[method]} TADs")
    for i in TADs:
        tad_bgn = i[0] / res - bgn
        tad_end = i[1] / res - bgn
        plt.plot([tad_bgn, tad_end], [tad_bgn, tad_bgn], color=tads_color, linewidth=5)
        plt.plot([tad_end, tad_end], [tad_bgn, tad_end], color=tads_color, linewidth=5)
    plt.savefig(f"{method}_tads_noise12_sim1.png")
```
